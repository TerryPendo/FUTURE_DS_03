```python
df=data.drop(columns=['Student ID'])
average_ratings=df.mean()

plt.figure(figsize=(18,5))
sns.barplot(y=average_ratings.index,x=average_ratings.values,hue= average_ratings.index, palette="viridis", legend = False)
plt.xlabel('Average Rating')
plt.ylabel('Feedback Category')
plt.title('Average Feedback Ratings')
plt.show();
```

## Average ratings  key takeaways
- **Highest-rated aspects**
  - *Well versed with the subject*  students feel the instructor is knowledgeable.
  - *Explains concepts in an understandable way*  clarity of instruction is strong.
  - *Use of presentations*  visual aids are valued by many students.
- **Lowest-rated aspect**
  - *Degree of difficulty of assignments*  assignments are consistently perceived as too difficult.


## **Distribution of ratings in a Histogram**

```python
plt.figure(figsize=(12,8))
df[average_ratings.index].hist(
    figsize=(12,8),
    xlabelsize=10,
    ylabelsize= 10,
    color=sns.color_palette("husl")[2],
    edgecolor="black",
    grid=True
)
plt.tight_layout(pad = 3.0, h_pad= 2.0, w_pad= 2.0)
plt.xticks(rotation = 45,ha = 'right', fontsize= 6)
plt.title("Distribution of Ratings", fontsize=15)
plt.show();
```

### Histogram Highlights
- Ratings across most categories are right-skewed  students generally give high scores.
- Assignment difficulty shows a low, narrow cluster  strong agreement that assignments are hard.
- Presentations show a wide spread  inconsistent student experience.
- Course recommendations cluster around the middle, matching the high 81% neutral sentiment.


## **Correlation Heatmap of Feedback Ratings**

```python
plt.figure(figsize=(8,7))
sns.heatmap(
    df[average_ratings.index].corr(),
    annot=True,
    cmap="crest",
    linewidths=0.2,
    square=True,
    cbar_kws={"shrink": .8}
)

plt.title("Correlation Heatmap of Feedback Ratings", fontsize=16)
plt.show()
```

## Correlation summary (heatmap)
- Correlations between ratings are **generally weak**, indicating that no single factor fully explains course recommendations.
- *Explains concepts well* shows the strongest (though still small) positive relation with **course recommendations**  clarity appears to matter more than other single factors.

## **Boxplots of Student Feedback**

```python
plt.figure(figsize=(10,4))
sns.boxplot(
    data=df[average_ratings.index],
    palette="rocket",
    width=0.6,
    linewidth=1.2
)

plt.title("Boxplot of Student Feedback Ratings", fontsize=16)
plt.xticks(rotation=45, ha='right')
sns.despine()
plt.show()
```

## Boxplot insights (variation and consistency)
- **Largest variation:** *Use of presentations*  students disagree on presentation usefulness; quality/consistency may vary across sessions.
- **Most consistent:** *Degree of difficulty of assignments*  students broadly agree assignments are difficult; this is a consistent pain point.


## **Overall Sentiment Score Per Student**

```python
sentiment_df = df.copy()

# Compute maximum possible rating per column
max_values = sentiment_df[average_ratings.index].max()

# Convert ratings to percentage of their max scale
percent_scores = sentiment_df[average_ratings.index].divide(max_values) * 100

# Compute each student's overall percentage satisfaction
sentiment_df["satisfaction_percent"] = percent_scores.mean(axis=1)

```

```python
def classify(score):
    if score >= 70:
        return "Positive"
    elif score >= 40:
        return "Neutral"
    else:
        return "Negative"

sentiment_df["sentiment"] = sentiment_df["satisfaction_percent"].apply(classify)
```

## **Pie Chart of Sentiments**

```python
plt.figure(figsize=(5,4))
sentiment_df["sentiment"].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=sns.color_palette("pastel"),
    startangle=140
)
plt.ylabel("")
plt.title("Overall Sentiment Distribution")
plt.show()
```

## Sentiment (scaled numeric sentiment)
- **Positive:** 17.4%  
- **Neutral:** 81.8%  
- **Negative:** 0.8%  
- Interpretation:
  - Most students are *neutral* (average/indifferent) rather than strongly positive or negative.
  - Low negative percentage suggests few students are actively dissatisfied.
  - Neutral majority suggests opportunity to convert neutrals to positives via targeted improvements.

## **Sentiment Count Bar Chart**

```python
sns.countplot(data=sentiment_df, x="sentiment",hue = "sentiment", palette="viridis", legend = False)
plt.title("Sentiment Count")
plt.xlabel("Sentiment Category")
plt.ylabel("Number of Students")
plt.show()
```
